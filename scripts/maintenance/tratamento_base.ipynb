```python
# ...existing code...
import numpy as np

def split_df(df, n=4, shuffle=False, random_state=None):
    """
    Divide `df` em `n` partes. Se shuffle=True, embaralha antes de dividir.
    Usa numpy.array_split para distribuir linhas de modo que os tamanhos
    fiquem o mais parecidos possível (diferença ≤ 1).
    """
    if shuffle:
        df = df.sample(frac=1, random_state=random_state).reset_index(drop=True)
    parts = np.array_split(df, n)
    return parts

# Exemplo de uso:
parts = split_df(cliente, n=4, shuffle=False)
df1, df2, df3, df4 = parts
print([len(p) for p in parts])  # Mostra os tamanhos

# Se quiser forçar pedaços exatamente iguais (truncando o resto, caso não seja múltiplo de 4):
rows = len(cliente)
group_size = rows // 4
cliente_trunc = cliente.iloc[:group_size * 4].reset_index(drop=True)
exact_parts = [cliente_trunc.iloc[i*group_size:(i+1)*group_size].reset_index(drop=True) for i in range(4)]
print([len(p) for p in exact_parts])
```

```python
# ...existing code...
def _find_name_col(df):
    # tenta identificar a coluna de nome do cliente
    candidates = ['nome', 'nome_cliente', 'cliente', 'name', 'nome_completo', 'full_name']
    for c in candidates:
        if c in df.columns:
            return c
    # fallback: primeira coluna objeto que não pareça ser ID
    for c in df.select_dtypes(include=['object', 'string']).columns:
        if 'id' not in c.lower():
            return c
    return None

def map_historico_by_clients(df_historico, clients_df, id_cols=['id_cliente', 'id_unico']):
    """
    Retorna uma cópia de df_historico onde id_cols são substituídas pelos nomes
    presentes em clients_df. Quando não houver correspondência, mantém o valor original.
    """
    name_col = _find_name_col(clients_df)
    df_out = df_historico.copy()
    if name_col is None:
        raise ValueError("Não foi encontrada coluna de nome no DataFrame de clientes.")
    for id_col in id_cols:
        if id_col in df_out.columns and id_col in clients_df.columns:
            # align tipos para merge
            df_out[id_col] = df_out[id_col].astype(str)
            clients_df[id_col] = clients_df[id_col].astype(str)
            # nome temporário para evitar sobrescrita
            tmp_name_col = f"{name_col}_tmp"
            merge_df = clients_df[[id_col, name_col]].rename(columns={name_col: tmp_name_col})
            df_out = df_out.merge(merge_df, how='left', on=id_col)
            # substitui id_col pelo nome quando houver correspondência
            df_out[id_col] = df_out[tmp_name_col].fillna(df_out[id_col])
            df_out.drop(columns=[tmp_name_col], inplace=True)
    return df_out

# Cria lista de históricos mapeados para cada subset
parts = split_df(cliente, n=4, shuffle=False)  # garante que `parts` exista
df1, df2, df3, df4 = parts

hist_parts = []
for i, part in enumerate([df1, df2, df3, df4], start=1):
    mapped = map_historico_by_clients(df_historico, part)
    hist_parts.append(mapped)
    # salvar opcionalmente
    mapped.to_excel(f"df_historico_mapped_{i}.xlsx", index=False)
    print(f"df_historico_mapped_{i}.xlsx gerado — linhas: {len(mapped)}")

# Atribui variáveis para uso posterior no notebook
df_historico1, df_historico2, df_historico3, df_historico4 = hist_parts
```

```python
# ...existing code...
import random
import numpy as np
import pandas as pd

def _find_date_col(df):
    candidates = ['data', 'date', 'dia', 'dt']
    for c in candidates:
        if c in df.columns:
            return c
    # procura colunas datetime-like
    for c in df.columns:
        if pd.api.types.is_datetime64_any_dtype(df[c]):
            return c
        # tenta parse para avaliar
        try:
            if pd.to_datetime(df[c], errors='coerce').notna().sum() > 0:
                return c
        except Exception:
            continue
    return None

def random_dates_between(start, end, n, seed=None):
    start = pd.to_datetime(start)
    end = pd.to_datetime(end)
    if seed is not None:
        np.random.seed(seed)
    days = (end - start).days
    if days <= 0:
        return pd.to_datetime([start] * n)
    rand_days = np.random.randint(0, days + 1, size=n)
    return start + pd.to_timedelta(rand_days, unit='d')

def augment_historico_by_random(df_historico, clients_df, n_new_per_row=1,
                                date_col=None, name_col=None,
                                name_target_cols=None,
                                date_randomization='full',
                                start_date=None, end_date=None, seed=None):
    """
    Retorna df_historico com novas linhas copiadas/alteradas:
    - name_col: coluna de nomes em clients_df (detecta automaticamente se None)
    - name_target_cols: colunas em df_historico a serem substituídas pelo nome aleatório (padrao: ['id_cliente','id_unico'] existentes)
    - date_col: coluna de data no df_historico (detecta automaticamente se None)
    - date_randomization: 'full' => data aleatória entre start/end; 'day' => mantem mes/ano e randomiza dia
    - n_new_per_row: quantas linhas novas criar por linha original
    """
    if seed is not None:
        random.seed(seed)
        np.random.seed(seed)

    if name_col is None:
        name_col = _find_name_col(clients_df)
    if name_col is None:
        raise ValueError("Não foi possível detectar coluna de nome em clients_df.")

    if date_col is None:
        date_col = _find_date_col(df_historico)

    # lista de nomes disponíveis no subset de clientes
    names = clients_df[name_col].dropna().astype(str).unique().tolist()
    if len(names) == 0:
        raise ValueError("Nenhum nome encontrado no clients_df.")

    # colunas no historico que receberão o nome aleatório (se existirem)
    if name_target_cols is None:
        name_target_cols = [c for c in ['id_cliente', 'id_unico'] if c in df_historico.columns]

    df_new_rows = []
    # define intervalo de datas para randomização "full"
    if date_randomization == 'full' and date_col is not None:
        s = pd.to_datetime(start_date) if start_date is not None else pd.to_datetime(df_historico[date_col].min())
        e = pd.to_datetime(end_date) if end_date is not None else pd.to_datetime(df_historico[date_col].max())
        if pd.isna(s): s = pd.Timestamp.now() - pd.Timedelta(days=365)
        if pd.isna(e): e = pd.Timestamp.now()

    for _, row in df_historico.iterrows():
        for _ in range(n_new_per_row):
            new_row = row.copy()
            chosen_name = random.choice(names)
            # aplica nome nas colunas de destino
            for c in name_target_cols:
                new_row[c] = chosen_name

            # randomiza a data
            if date_col is not None and date_col in df_historico.columns:
                if date_randomization == 'full':
                    new_date = random_dates_between(s, e, 1, seed=seed)[0]
                elif date_randomization == 'day':
                    try:
                        orig_date = pd.to_datetime(row[date_col])
                        year = orig_date.year
                        month = orig_date.month
                        last_day = pd.Period(f'{year}-{month:02d}').days_in_month
                        day = random.randint(1, last_day)
                        new_date = pd.Timestamp(year=year, month=month, day=day)
                    except Exception:
                        new_date = orig_date if 'orig_date' in locals() else pd.Timestamp.now()
                else:
                    new_date = pd.to_datetime(row[date_col])
                new_row[date_col] = new_date
            df_new_rows.append(new_row)

    df_new = pd.DataFrame(df_new_rows)
    df_out = pd.concat([df_historico, df_new], ignore_index=True)
    return df_out

# Exemplo de uso para cada subset (cria 2 novos registros por linha):
df_historico1_aug = augment_historico_by_random(df_historico1, df1, n_new_per_row=2, date_randomization='full', seed=42)
df_historico2_aug = augment_historico_by_random(df_historico2, df2, n_new_per_row=2, date_randomization='full', seed=42)
df_historico3_aug = augment_historico_by_random(df_historico3, df3, n_new_per_row=2, date_randomization='full', seed=42)
df_historico4_aug = augment_historico_by_random(df_historico4, df4, n_new_per_row=2, date_randomization='full', seed=42)

# Salva e mostra contagens
df_historico1_aug.to_excel("df_historico1_aug.xlsx", index=False)
df_historico2_aug.to_excel("df_historico2_aug.xlsx", index=False)
df_historico3_aug.to_excel("df_historico3_aug.xlsx", index=False)
df_historico4_aug.to_excel("df_historico4_aug.xlsx", index=False)

print("linhas originais / depois (1):", len(df_historico1), "/", len(df_historico1_aug))
print("linhas originais / depois (2):", len(df_historico2), "/", len(df_historico2_aug))
print("linhas originais / depois (3):", len(df_historico3), "/", len(df_historico3_aug))
print("linhas originais / depois (4):", len(df_historico4), "/", len(df_historico4_aug))
```
